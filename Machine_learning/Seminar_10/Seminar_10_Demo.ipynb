```python
#coding=utf-8
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors
import numpy as np
```

## In this demonstration we will compare three methods: LDA, QDA and Logistic Regression

```python
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA, QuadraticDiscriminantAnalysis as QDA
from sklearn.model_selection import train_test_split
```

```python
def test_methods(X_train, X_test, y_train, y_test):
    methods = [LDA(), QDA(), LogisticRegression()]
    method_names = ['LDA', 'QDA', 'LR']
    accuracies = []
    roc_aucs = []
    for i in range(len(methods)):
        methods[i].fit(X_train, y_train)
        y_pred = methods[i].predict(X_test)
        y_prob = methods[i].predict_proba(X_test)[:, 1]
        accuracies.append(metrics.accuracy_score(y_test, y_pred))
        roc_aucs.append(metrics.roc_auc_score(y_test, y_prob))
    print('accuracy:')
    for i in range(len(methods)):
        print(method_names[i], accuracies[i])
    print('\nroc_auc:')
    for i in range(len(methods)):
        print(method_names[i], roc_aucs[i])
```

```python
def plot_decision_boundary(class_1, class_2):
    plt.subplot(1, 3, 1)
    plt.title('LogisticRegression')
    plt.scatter(class_1[:, 0], class_1[:, 1], s = 5, color = 'r')
    plt.scatter(class_2[:, 0], class_2[:, 1], s = 5, color = 'b')
    plt.axis('equal');
    x_min, x_max = plt.xlim()
    y_min, y_max = plt.ylim()
    nx, ny = 100, 100
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, nx),
                         np.linspace(y_min, y_max, ny))
    lr = LogisticRegression()
    lr.fit(X, y)
    Z = lr.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    Z = Z[:, 1].reshape(xx.shape)
    plt.pcolormesh(xx, yy, Z, cmap=cm.coolwarm_r, norm=colors.Normalize(-0.5, 1.5))
    plt.contour(xx, yy, Z, [0.5], linewidths=2., colors='k')
    plt.scatter(class_1[:, 0], class_1[:, 1], s = 5, color = 'r')
    plt.scatter(class_2[:, 0], class_2[:, 1], s = 5, color = 'b')

    plt.subplot(1, 3, 2)
    plt.title('LDA')
    lda = LDA()
    lda.fit(X, y)
    Z = lda.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    Z = Z[:, 1].reshape(xx.shape)
    plt.pcolormesh(xx, yy, Z, cmap=cm.coolwarm_r, norm=colors.Normalize(-0.5, 1.5))
    plt.contour(xx, yy, Z, [0.5], linewidths=2., colors='k')
    plt.scatter(class_1[:, 0], class_1[:, 1], s = 5, color = 'r')
    plt.scatter(class_2[:, 0], class_2[:, 1], s = 5, color = 'b')
    plt.axis('equal');
    plt.subplot(1, 3, 3)
    plt.title('QDA')
    qda = QDA()
    qda.fit(X, y)
    Z = qda.predict_proba(np.c_[xx.ravel(), yy.ravel()])
    Z = Z[:, 1].reshape(xx.shape)
    plt.pcolormesh(xx, yy, Z, cmap=cm.coolwarm_r, norm=colors.Normalize(-0.5, 1.5))
    plt.contour(xx, yy, Z, [0.5], linewidths=2., colors='k')
    plt.scatter(class_1[:, 0], class_1[:, 1], s = 5, color = 'r')
    plt.scatter(class_2[:, 0], class_2[:, 1], s = 5, color = 'b')
    plt.axis('equal');
```

## Case 1: 2d normally distributed classes with similar covariances
In the simpliest case all methods should work almost equally

```python
# generate sample
np.random.seed(0)
N = 100
d = 2
mu = np.zeros(d)
sigma = np.eye(d)
offset = 2
class_1 = np.random.multivariate_normal(mu, sigma, N)
class_2 = np.random.multivariate_normal(mu + offset, sigma, N)
X = np.vstack((class_1, class_2))
y = np.hstack((np.zeros(N), np.ones(N)))

# plot sample
plt.figure(figsize = (15, 5))
plot_decision_boundary(class_1, class_2)

# test methods
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
test_methods(X_train, X_test, y_train, y_test)
```

## Case 2: 2d normally distributed classes with different covariances
QDA outperforms LDA and Logistic Regression since its decision boundary is more sophisticated.

```python
# generate sample
np.random.seed(0)
N = 500
d = 2
mu = np.zeros(d)
sigma_1 = np.eye(d)
sigma_1[0,0] = 10
sigma_2 = np.eye(d)
sigma_2[1,1] = 10
offset = 2.0
class_1 = np.random.multivariate_normal(mu, sigma_1, N)
class_2 = np.random.multivariate_normal(mu + offset, sigma_2, N)
X = np.vstack((class_1, class_2))
y = np.hstack((np.zeros(N), np.ones(N)))

# plot sample
plt.figure(figsize = (15, 5))
plot_decision_boundary(class_1, class_2)
# test methods
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
test_methods(X_train, X_test, y_train, y_test)
```

## Case 3: normally distributed classes with ill-conditioned similar covariances (linearly dependent features)
Here one can notice that LDA perfoms better than QDA since it has increased numerical stability due to sigma equality assumption. With the growth of the sample size this difference vanishes. 

```python
# generate sample
np.random.seed(2)
N = 100
d = 20
mu = np.zeros(d)
alpha = 0.01
sigma = alpha * np.eye(d) + (1 - alpha)
offset = np.zeros(d)
offset[1] = 0.2
class_1 = np.random.multivariate_normal(mu, sigma, N)
class_2 = np.random.multivariate_normal(mu + offset, sigma, N)
X = np.vstack((class_1, class_2))
y = np.hstack((np.zeros(N), np.ones(N)))

# plot sample
plt.figure(figsize = (5, 5))
plt.scatter(class_1[:, 0], class_1[:, 1], s = 5, color = 'r')
plt.scatter(class_2[:, 0], class_2[:, 1], s = 5, color = 'b')
plt.axis('equal');

# test methods
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
test_methods(X_train, X_test, y_train, y_test)
```

## Case 4: 2d normally distributed classes with gross outliers
Logistic Regession is more robust.

```python
# generate sample
np.random.seed(2)
N = 300
d = 2
mu = np.zeros(d)
alpha = 0.1
sigma = alpha * np.eye(d) + (1 - alpha)
offset = np.zeros(d)
offset[1] = 2.0
outliers_1 = np.random.multivariate_normal(mu + 5*offset, np.eye(d)*10, 10)
outliers_2 = np.random.multivariate_normal(mu + 5*offset, np.eye(d)*10, 10)
class_1 = np.vstack((np.random.multivariate_normal(mu, sigma, N)))
class_2 = np.vstack((np.random.multivariate_normal(mu + offset, sigma * 5, N)))
X = np.vstack((class_1, class_2))
y = np.hstack((np.zeros(class_1.shape[0]), np.ones(class_2.shape[0])))

# plot sample
plt.figure(figsize = (5, 5))
plt.scatter(class_1[:, 0], class_1[:, 1], s = 5, color = 'r')
plt.scatter(class_2[:, 0], class_2[:, 1], s = 5, color = 'b')
plt.scatter(outliers_1[:, 0], outliers_1[:, 1], s = 50, marker = 'x', color = 'r')
plt.scatter(outliers_2[:, 0], outliers_2[:, 1], s = 50, marker = 'x', color = 'b')
plt.axis('equal');

# test methods
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
X_train = np.vstack((X_train, outliers_1, outliers_2))
y_train = np.hstack((y_train, np.zeros(len(outliers_1)), np.ones(len(outliers_2))))
test_methods(X_train, X_test, y_train, y_test)
```

## Case 5: 2d non-normally distributed classes
Logistic Regression outperforms normal discriminant analysis

```python
# generate sample
np.random.seed(4)
N = 5000
d = 2
mu = np.zeros(d)
sigma_1 = np.eye(d)
sigma_1[0,0] = 10
sigma_2 = np.eye(d)
sigma_2[1,1] = 10
class_1 = np.random.multivariate_normal(mu, sigma_1, N)
class_1 = class_1[class_1[:, 1] > 0]
class_2 = np.random.multivariate_normal(mu, sigma_2, N)
class_2 = class_2[class_2[:, 1] < 0]
X = np.vstack((class_1, class_2))
y = np.hstack((np.zeros(class_1.shape[0]), np.ones(class_2.shape[0])))

# plot sample
plt.figure(figsize = (15, 5))
plot_decision_boundary(class_1, class_2)

# test methods
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
test_methods(X_train, X_test, y_train, y_test)
```

## Case 6: 2d non-normally distributed classes with non-linear separator
QDA outperforms LDA and Logistic Regression

```python
# generate sample
np.random.seed(4)
N = 500
d = 2
mu = np.zeros(d)
sigma_1 = np.eye(d)
sigma_2 = np.eye(d)
offset = 5
class_1 = np.vstack((np.random.multivariate_normal(mu, sigma_1, N), np.random.multivariate_normal(mu + 2*offset, sigma_1, N)))
class_2 = np.random.multivariate_normal(mu + offset + np.array([0, 1]), sigma_2, N)
X = np.vstack((class_1, class_2))
y = np.hstack((np.zeros(class_1.shape[0]), np.ones(class_2.shape[0])))

# plot sample
plt.figure(figsize = (15, 5))
plot_decision_boundary(class_1, class_2)
# test methods
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
test_methods(X_train, X_test, y_train, y_test)
```
