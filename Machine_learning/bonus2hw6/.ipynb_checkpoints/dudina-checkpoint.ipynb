* Make conclusion

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools as it

from matplotlib import cm
from sklearn.preprocessing import scale
from sklearn.linear_model import LinearRegression 
from sklearn.decomposition import PCA
from sklearn.manifold import LocallyLinearEmbedding, MDS, SpectralEmbedding, TSNE
from sklearn.model_selection import KFold,  cross_val_score
from statsmodels.regression.linear_model import OLS, RegressionResults

%matplotlib inline
```

```python
data = pd.read_csv(open('prostate.csv'))
```

```python
columns = data.columns
```

```python
X, y = np.array(data[columns[:8]]), np.array(data[columns[8]])
```

```python
subsets = []
for i in range(1, 9):
    subset = np.array(list(it.combinations(columns[:8], i)))
    subsets.append((subset))
subsets1 = []
for i in range(len(subsets)):
     subsets1=subsets1 + subsets[i].tolist()
subsets1 = np.array(subsets1)
s= len(subsets1)
print (s)
```

```python
AIC=[]
BIC=[]
five_fold=[]
ten_fold=[]

for i in range(s):
    X = np.asarray(data[subsets1[i]])
    clf= OLS(y, X)
    CLF = clf.fit()
    AIC.append(CLF.aic)
    BIC.append(CLF.bic)
    five_fold.append(-cross_val_score(LinearRegression(), X, y, scoring='neg_mean_squared_error', cv=5).mean())
    ten_fold.append(-cross_val_score(LinearRegression(), X, y, scoring='neg_mean_squared_error', cv=10).mean())
print (min(AIC))
```

```python
from __future__ import print_function
print ('Best AIC score', min(AIC))
print ('Best AIC subset', subsets1[np.argmin(BIC)])
print ('Best BIC score', min(BIC))
print ('Best BIC subset', subsets1[np.argmin(BIC)])
print ('Best five_fold score', min(five_fold))
print ('Best five_fold subset', subsets1[np.argmin(five_fold)])
print ('Best ten_fold score', min(ten_fold))
print ('Best ten_fold subset', subsets1[np.argmin(ten_fold)])
```

```python
n=2
clf = PCA(n_components=n)
XPCA = clf.fit_transform(X)
```

```python
plt.figure(figsize=(10, 7))
cc = [cm.Accent((y_) / max(y)) for y_ in y]
plt.scatter(XPCA[:,0], XPCA[:,1],c=cc,  s=30)
plt.title("PCA ")
```

```python
n=2
clf = LocallyLinearEmbedding(n_components=n)
XLLE = clf.fit_transform(X)
plt.figure(figsize=(10, 7))
cc = [cm.Accent((y_) / max(y)) for y_ in y]
plt.scatter(XLLE[:,0], XLLE[:,1],c=cc,  s=30)
plt.title("LLE ")
```

```python
n=2
clf = MDS(n_components=n)
XMDS = clf.fit_transform(X)
plt.figure(figsize=(10, 7))
cc = [cm.Accent((y_) / max(y)) for y_ in y]
plt.scatter(XMDS[:,0], XMDS[:,1],c=cc,  s=30)
plt.title("MDS")
```

```python
n=2
clf = SpectralEmbedding(n_components=n)
XSpEm = clf.fit_transform(X)
plt.figure(figsize=(10, 7))
cc = [cm.Accent((y_) / max(y)) for y_ in y]
plt.scatter(XSpEm[:,0], XSpEm[:,1],c=cc,  s=30)
plt.title("Laplacian Eigen maps")
```

```python
n=2
clf = TSNE(n_components=n)
tSNE = clf.fit_transform(X)
plt.figure(figsize=(10, 7))
cc = [cm.Accent((y_) / max(y)) for y_ in y]
plt.scatter(tSNE[:,0], tSNE[:,1],c=cc,  s=30)
plt.title("tSNE")
```
