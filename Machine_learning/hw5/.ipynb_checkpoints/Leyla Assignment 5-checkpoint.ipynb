```python
X_plot = np.linspace(-5, 10, 1000)[:, np.newaxis]
X = np.array(glass['RI']).reshape(-1,1)
```

```python
cl_data = np.loadtxt("clustering.txt")
labels = np.loadtxt("clustering_labels.txt")
```

```python
#visualization of all clusters without colours
plt.figure(figsize=(10,7))
plt.scatter(cl_data[:,0], cl_data[:,1], s=10)
```

```python
def plot_clustering(method):
    num_clasters = [2, 15, 30]
    plt.figure(figsize = (10, 5))
    for num_cluster in num_clasters:
        method.fit(cl_data)
        predicted_clusters = method.predict(cl_data)
        C = len(set(predicted_clusters))
        colors = [cm.Vega10(c) for c in predicted_clusters]
        plt.scatter(cl_data[:, 0], cl_data[:, 1], color = colors, edgecolors='none', s = 3)
        plt.title("%s" %method)
```

## K means

```python
#k means for 3 clusters 
plot_clustering(cluster.KMeans(n_clusters=3))
#k means for 15 clusters (right number)
plot_clustering(cluster.KMeans(n_clusters=15))
#k means for 30 clusters 
plot_clustering(cluster.KMeans(n_clusters=30))
```

## Gaussian Mixture

```python
#Gaussian Mixture for 3 clusters 
plot_clustering(GaussianMixture(n_components=3))
#Gaussian Mixture for 15 clusters (right number)
plot_clustering(GaussianMixture(n_components=15))
#k means for 30 clusters 
plot_clustering(GaussianMixture(n_components=30))
```

## Hierarchical Ward clustering

```python
plt.figure(figsize = (10, 7))
num_clusters = [3, 15, 30]
method = cluster.AgglomerativeClustering(n_clusters=15, linkage='ward')
method.fit(cl_data)
predicted_clusters = method.labels_
C = len(set(predicted_clusters))
colors = [cm.Vega10(c) for c in predicted_clusters]
plt.scatter(cl_data[:, 0], cl_data[:, 1], color = colors, edgecolors='none', s = 3)
```
