```python
np.random.seed(9) #9, 7
shuffled_inds = np.random.choice(len(digits['target']), len(digits['target']), replace=False)
test_size = 500
test_inds = shuffled_inds[:test_size]
train_inds = shuffled_inds[test_size:]
```

```python
from sklearn.semi_supervised import LabelPropagation
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
from scipy import stats
```

```python
import warnings
warnings.filterwarnings('ignore')
```

# Naive strategy
* Start with a pool of unlabeled data $S \subset  X$
* Pick a few points from $S$ at random and get their labels
* Repeat:
    * Fit a classifier $h \in H$ to the labels seen so far
    * Query the unlabeled point in $S$ closest to the boundary of $h$ (or most uncertain, or most likely to decrease overall uncertainty,...)


```python
initial_size = 30
iter_addition = 10

X_test = digits['data'][test_inds]
y_test = digits['target'][test_inds]

active_inds = train_inds[:initial_size]
accuracies = []
active_accuracies = []
active_sizes = []
for active_iter in range(10):
    active_size = initial_size + active_iter * iter_addition
    
    X_train = np.vstack((digits['data'][train_inds[:active_size]], digits['data'][train_inds[active_size:]]))
    y_train = np.hstack((digits['target'][train_inds[:active_size]], np.zeros(len(train_inds) - active_size) - 1))
    grid_search = LabelPropagation(kernel='knn', n_neighbors=15, alpha=0.9, max_iter=5)    
    grid_search.fit(X_train, y_train)
    
    accuracies.append(accuracy_score(grid_search.predict(X_test), y_test))
    
    other_inds = np.array(list(set(train_inds) - set(active_inds)))
    X_train_active = np.vstack((digits['data'][active_inds], digits['data'][other_inds]))
    y_train_active = np.hstack((digits['target'][active_inds], np.zeros(len(other_inds)) - 1))
    grid_search_active = LabelPropagation(kernel='knn', n_neighbors=15, alpha=0.9, max_iter=5)
    grid_search_active.fit(X_train_active, y_train_active)
    active_accuracies.append(accuracy_score(grid_search_active.predict(X_test), y_test))    
    active_sizes.append(active_size)
    
    # add the most uncertain elements to the active set 
    y_prob = grid_search_active.label_distributions_
    pred_entropies = stats.distributions.entropy(y_prob.T)
    uncertain_inds = np.argsort(pred_entropies)[::-1]

    i = 0
    selected = 0
    plt.figure(figsize = (iter_addition, 2))
    while selected < iter_addition:
        if train_inds[uncertain_inds[i]] not in active_inds:
            active_inds = np.append(active_inds, train_inds[uncertain_inds[i]])
            selected += 1
            plt.subplot(1, iter_addition, selected)
            cur_predicred_class = grid_search_active.transduction_[uncertain_inds[i]]
            cur_true_class = digits['target'][train_inds[uncertain_inds[i]]]
            if cur_predicred_class == cur_true_class:
                plt.title(cur_true_class)
            else:
                plt.title(str(int(cur_predicred_class)) + ' # ' + str(cur_true_class))
            plt.imshow(digits['images'][train_inds[uncertain_inds[i]]], cmap=plt.cm.gray, interpolation='nearest')
            plt.xticks(())
            plt.yticks(())
        i += 1
```

```python
plt.plot(active_sizes, accuracies, marker = '.', color = 'blue', label = 'random')
plt.plot(active_sizes, active_accuracies, marker = '.', color = 'red', label = 'active')
plt.xlabel('active train size')
plt.ylabel('accuracy')
plt.legend(loc = 4)
```

# exploration-exploitation tradeoff

```python
initial_size = 30
iter_addition = 10

X_test = digits['data'][test_inds]
y_test = digits['target'][test_inds]

active_inds = train_inds[:initial_size]
active_accuracies = []
active_sizes = []
for active_iter in range(10):
    active_size = initial_size + active_iter * iter_addition
    
    other_inds = np.array(list(set(train_inds) - set(active_inds)))
    X_train_active = np.vstack((digits['data'][active_inds], digits['data'][other_inds]))
    y_train_active = np.hstack((digits['target'][active_inds], np.zeros(len(other_inds)) - 1))
    grid_search_active = LabelPropagation(kernel='knn', n_neighbors=15, alpha=0.9, max_iter=5)
    grid_search_active.fit(X_train_active, y_train_active)
    active_accuracies.append(accuracy_score(grid_search_active.predict(X_test), y_test))    
    active_sizes.append(active_size)
    
    # add the most uncertain elements to the active set 
    y_prob = grid_search_active.label_distributions_
    pred_entropies = stats.distributions.entropy(y_prob.T)
    uncertain_inds = np.argsort(pred_entropies)[::-1]

    #uncertain_inds = np.argsort(np.max(y_prob, axis = 1))
    i = 0
    selected = 0
    plt.figure(figsize = (iter_addition, 2))
    while selected < iter_addition/2:
        if train_inds[uncertain_inds[i]] not in active_inds:
            active_inds = np.append(active_inds, train_inds[uncertain_inds[i]])
            selected += 1
            plt.subplot(2, iter_addition, selected)
            cur_predicred_class = grid_search_active.transduction_[uncertain_inds[i]]
            cur_true_class = digits['target'][train_inds[uncertain_inds[i]]]
            if cur_predicred_class == cur_true_class:
                plt.title(cur_true_class)
            else:
                plt.title(str(int(cur_predicred_class)) + ' # ' + str(cur_true_class))
            plt.imshow(digits['images'][train_inds[uncertain_inds[i]]], cmap=plt.cm.gray, interpolation='nearest')
            plt.xticks(())
            plt.yticks(())
            plt.subplot(2, iter_addition, selected + iter_addition)
            plt.bar(range(len(y_prob[uncertain_inds[i]])), y_prob[uncertain_inds[i]])
            plt.xticks(())
            plt.yticks(())
        i += 1
    while selected < iter_addition:
        i = np.random.randint(active_size + iter_addition)
        if train_inds[i] not in active_inds:
            active_inds = np.append(active_inds, train_inds[i])
            selected += 1
            plt.subplot(2, iter_addition, selected)
            cur_predicred_class = grid_search_active.transduction_[uncertain_inds[i]]
            cur_true_class = digits['target'][train_inds[i]]
            if cur_predicred_class == cur_true_class:
                plt.title(cur_true_class)
            else:
                plt.title(str(int(cur_predicred_class)) + ' # ' + str(cur_true_class))
            plt.imshow(digits['images'][train_inds[i]], cmap=plt.cm.gray, interpolation='nearest')
            plt.xticks(())
            plt.yticks(())
            plt.subplot(2, iter_addition, selected + iter_addition)
            plt.bar(range(len(y_prob[i])), y_prob[i], color = 'orange')
            plt.xticks(())
            plt.yticks(())
```

```python
plt.plot(active_sizes, accuracies, marker = '.', color = 'blue', label = 'random')
plt.plot(active_sizes, active_accuracies, marker = '.', color = 'red', label = 'active')
plt.xlabel('active train size')
plt.ylabel('accuracy')
plt.legend(loc = 4)
```
