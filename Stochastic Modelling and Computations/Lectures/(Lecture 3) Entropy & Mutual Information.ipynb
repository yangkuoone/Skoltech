```julia
using PyPlot
using Distributions
```

```julia
p0=[0.05 0.15; 0.05 0.75] # example of a joint disctribution of two variables

marg1(p0)=sum(p0,1)
marg2(p0)=sum(p0,2)
cond1(p0)=p0./[marg1(p0); marg1(p0)]
cond2(p0)=p0./[marg2(p0)  marg2(p0)]

show(cond1(p0))
cond2(p0)
```

```julia
pb = linspace(0, 1, 100)
PyPlot.plot(pb,-pb.*log(pb)-(1-pb).*log(1-pb))
PyPlot.title("Entropy of the Bernoulli Distribution")
```

```julia
ent(p)=sum(-p.*log(p))

p1=[1/5 1/5 1/5 1/5 1/5] 
p2=[1/2 1/3 1/6]
p3=[1/2 1/10 1/10 1/10 1/10 1/10]

[ent(p1)
ent(p2)
    ent(p3)]
```

```julia
mutual_inf(p)=-ent(p)+ent(sum(p,1))+ent(sum(p,2))

p_u=1/4*[1.1 0.9; 1 1]

[ent(p_u)
    mutual_inf(p_u)]
```

```julia
KL(p1,p2)=sum(p1.*log(p1./p2)) # The two distributions should have the same support

[KL(p0,p_u) KL(p_u,p0) KL(p0,p0) KL(p_u,p_u)]
```
