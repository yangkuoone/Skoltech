```python
lambd = 0.5
mu = 1 - lambd
initial_pos = 0
num_iter = 1000
positions = np.zeros(num_iter)

for it in range(1, num_iter):
    c = np.random.rand()
    if c <= lambd:
        positions[it] = positions[it - 1] + 1
    else:
        positions[it] = positions[it - 1] - 1
        
plt.title('Particle trajectory for lambda = %s' % lambd)
plt.plot(positions)
plt.show()
```
