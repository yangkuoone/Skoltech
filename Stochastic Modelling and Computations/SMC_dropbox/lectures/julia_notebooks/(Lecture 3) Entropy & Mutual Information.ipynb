```julia
p=[0.05 0.15; 0.05 0.75] # example of a joint disctribution of two variables

marg1(p)=sum(p,1)
marg2(p)=sum(p,2)
cond1(p)=p./[marg1(p); marg1(p)]
cond2(p)=p./[marg2(p)  marg2(p)]

show(cond1(p))
cond2(p)
```

```julia
using PyPlot

p = linspace(0, 1, 100)
PyPlot.plot(p,-p.*log(p)-(1-p).*log(1-p))
PyPlot.title("Entropy of the Bernoulli Distribution")
```

```julia
ent(p)=sum(-p.*log(p))

p1=[1/5 1/5 1/5 1/5 1/5] 
p2=[1/2 1/3 1/6]
p3=[1/2 1/10 1/10 1/10 1/10 1/10]

[ent(p1)
ent(p2)
    ent(p3)]
```

```julia
mutual_inf(p)=-ent(p)+ent(sum(p,1))+ent(sum(p,2))

p_u=1/4*[1.1 0.9; 1 1]

[ent(p_u)
    mutual_inf(p_u)]
```

```julia
KL(p1,p2)=sum(p1.*log(p1./p2)) # The two distributions should have the same support

[KL(p,p_u) KL(p_u,p) KL(p,p) KL(p_u,p_u)]
```
